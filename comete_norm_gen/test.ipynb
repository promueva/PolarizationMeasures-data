```python
import matplotlib.pyplot as plt

def plot_polarization_values(file_path):
    with open(file_path, 'r') as file:
        file.__next__()
        numbers = [float(line.strip()) for line in file]

    # Create a plot
    plt.figure(figsize=(10, 5))

    # Plot each number as a dot
    plt.plot(numbers, 'o', color='blue')

    # Add labels and title
    plt.ylabel('Value')
    plt.title('Unnormalized values of COMETE α=1.2 β=1')

    # Display the plot
    plt.show()
```

```python
plot_polarization_values("pop100/alpha1_beta1")
plot_polarization_values("pop100/alpha1.2_beta1")
plot_polarization_values("pop100/alpha2_beta1")
```

```python
def standarize(input_file_path, output_file_path):
    # Read numbers from the input text file
    with open(input_file_path, 'r') as infile:
        header = infile.__next__()
        numbers = [float(line.strip()) for line in infile]

    # Format each number to 10 decimal places
    formatted_numbers = [f"{num:.10f}" for num in numbers]

    # Write the formatted numbers to the output text file
    with open(output_file_path, 'w') as outfile:
        outfile.write(header)
        for num in formatted_numbers:
            outfile.write(f"{num}\n")

#standarize("alpha1_beta1", "10f/alpha1_beta1")
#standarize("alpha2_beta1", "10f/alpha2_beta1")
#standarize("alpha1.2_beta1","10f/alpha1.2_beta1")
```
