```python
import matplotlib.pyplot as plt

# File path
file_path = "alpha1_beta1_FINAL"

# Read numbers from text file
with open(file_path, 'r') as file:
    file.__next__()
    numbers = [float(line.strip()) for line in file]

# Create a plot
plt.figure(figsize=(10, 5))

# Plot each number as a dot
plt.plot(numbers, 'o', color='blue')

# Add labels and title
plt.xlabel('Histogram index')
plt.ylabel('Value')
plt.title('Unnormalized values of COMETE α=1 β=1')

# Display the plot
plt.show()
```
